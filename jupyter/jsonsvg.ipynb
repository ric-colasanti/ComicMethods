```python
import cv2
import json 
import numpy as np
from matplotlib import pyplot as plt

def loadJFrame(fileName):
    with open(fileName, 'r') as myfile:
        data=myfile.read()
    obj = json.loads(data)
    img2 = np.array(obj["img"])
    plt.imshow(img2)
    plt.show()
    for i in range (len(obj["Character"])):
        byMin = obj["Character"][i]["bbox"][0]
        byMax = obj["Character"][i]["bbox"][1]
        bxMin = obj["Character"][i]["bbox"][2]
        bxMax = obj["Character"][i]["bbox"][3]
        img3 = img2[byMin:byMax,bxMin:bxMax,::-1]
        plt.imshow(img3)
        plt.show()

    for i in range (len(obj["Balloon"])):
        byMin = obj["Balloon"][i]["bbox"][0]
        byMax = obj["Balloon"][i]["bbox"][1]
        bxMin = obj["Balloon"][i]["bbox"][2]
        bxMax = obj["Balloon"][i]["bbox"][3]
        img3 = img2[byMin:byMax,bxMin:bxMax,::]
        plt.imshow(img3)
        plt.show()
```
