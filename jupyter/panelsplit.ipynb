```python
import cv2
import json 
import os
from matplotlib import pyplot as plt

def boundingBox(string,buffer):
    s = string[1:-1]
    points = s.split(" ")
    xMax = 0
    yMax = 0
    yMin=1000000
    xMin=1000000
    for p in points:
        a = p.split(",")
        if int(a[0])<xMin:
            xMin = int(a[0])
        if int(a[0])>xMax:
            xMax = int(a[0])

        if int(a[1])<yMin:
            yMin = int(a[1])
        if int(a[1])>yMax:
            yMax = int(a[1])
    return xMin-buffer, xMax+buffer, yMin-buffer, yMax+buffer

    return True  


def within(axMin,axMax,ayMin,ayMax,bxMin,bxMax,byMin,byMax):
    if axMin>bxMin:
        return False
    if ayMin>byMin:
        return False
    if axMax<bxMax:
        return False
    if ayMax<byMax:
        return False
    return True  

def getPanel(obj,n,img):
    d =obj["Panel"][n]
    buffer = 5
    xMin,xMax,yMin,yMax = boundingBox(d['points'],buffer)
    img2 = img[yMin:yMax,xMin:xMax,::-1].tolist()
    return {"img":img2,"abspos":[xMin,xMax,yMin,yMax] }

def getBBox(ctype,obj,data):
    data[ctype]=[]
    for c in obj[ctype]:
        bxMin,bxMax,byMin,byMax = boundingBox(c['points'],0)
        if within(data["abspos"][0],data["abspos"][1],data["abspos"][2],data["abspos"][3],bxMin,bxMax,byMin,byMax):
            data[ctype].append({"bbox":[byMin-data["abspos"][2],byMax-data["abspos"][2],bxMin-data["abspos"][0],bxMax-data["abspos"][0]]})

   
```

```python
def extractFrames(filename):
    with open('/home/ric/Projects/Comic/Sandbox/'+filename+'.json', 'r') as myfile:
        data=myfile.read()
    print(filename,end=" ")
    img = cv2.imread('/home/ric/Projects/Comic/eBDtheque_database_v3/GT/'+filename+'.jpg')
    obj = json.loads(data)
    try:
        for i in range(len(obj["Panel"])):
            print(i,end= " ")
            data = getPanel(obj,i,img)
            getBBox("Character",obj,data)
            getBBox("Balloon",obj,data)
            with open('/home/ric/Projects/Comic/Sandbox/Frames/'+filename+'Frame'+str(i)+'.json', 'w') as outfile:
                json.dump(data, outfile)
        print(" OK")
    except:
        print(" Error")
```
